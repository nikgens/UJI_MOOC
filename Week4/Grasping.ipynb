## Grasping

Assuming that the robot is well positioned, the grasping task is straightforward:
    1. close fingers
    2. lift gripper

Not every task requires a lot of code! ;-)

```python
import packages.initialization
import pioneer3dx as p3dx
p3dx.init()
```

### Close fingers

```python
p3dx.gripper(0.05,0.0)
```

### Lift gripper

```python
p3dx.gripper(-0.05,0.0)
```
